```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

biomass = pd.read_csv('C:\\Users\\danan\\Documents\\UniversityYear2\\Dissertation\\Python\\Mozambique\\Good_Fields.csv', usecols=['date', 'BIOMASS_1'])
biomass.set_index('date',inplace=True,drop=True)

biomass.plot(legend=False)
plt.title('Biomass')
```

```python
ax = biomass.plot(legend=False)
ax.set_title('Rolling biomass over time')

biomass.rolling(window=10).mean().plot(ax=ax)
biomass.rolling(window=30).mean().plot(ax=ax)
biomass.rolling(window=50).mean().plot(ax=ax)
ax.legend(['Biomasss', 'm=10', 'm=30', 'm=50'])
```

```python
ax = biomass.plot(legend=False)
ax.set_title('Biomass over time')

biomass.rolling(window=10).var().plot(ax=ax)
biomass.rolling(window=30).var().plot(ax=ax)
biomass.rolling(window=50).var().plot(ax=ax)
ax.legend(['Biomass', 'm=10', 'm=30', 'm=50'])
```
